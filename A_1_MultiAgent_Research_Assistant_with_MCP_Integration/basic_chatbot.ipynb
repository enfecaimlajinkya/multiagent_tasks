```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langchain_openai import ChatOpenAI
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
print(groq_api_key)
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
from typing import TypedDict, Optional

class AgentState(TypedDict):
    question: str
    plan: Optional[str]
    research_result: Optional[str]
    final_answer: Optional[str]
```

Getting answers by using Duckduckgo

```python
user_question="Tadoba national park in which state and which districts?"
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

context = search_tool.invoke(user_question)

print(context)

prompt = f"""
You are a factual QA assistant.

Use ONLY the provided context to answer the question.
Do not use your own knowledge or hallucinate.
If the answer is not in the context, say "Not found".

Context:
{context}

Question:
{user_question}

Answer with a short factual response.
"""
```

Answers providing to Groq LLM

```python
response = llm.invoke(prompt)

print(response.content)
```

Planner Agent

```python
def planner_agent(state: AgentState):

    question = state["question"]

    prompt = f"""
You are a planning assistant.

Your job is to decide what needs to be researched.

User Question:
{question}

Write a simple research plan.
"""

    response = llm.invoke(prompt)

    return {"plan": response.content}
```
